```python
import pandas as pd

df = pd.read_csv('../data/application_train.csv')

# Tạo bản sao để test
df_fe = df[['TARGET', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'DAYS_BIRTH', 'DAYS_EMPLOYED']].copy()

# Feature 1: Tỷ lệ Nợ / Thu nhập (Khách vay quá tay?)
df_fe['PAYMENT_RATE'] = df_fe['AMT_ANNUITY'] / df_fe['AMT_INCOME_TOTAL']

# Feature 2: Tỷ lệ Khoản vay / Thu nhập
df_fe['CREDIT_TO_INCOME'] = df_fe['AMT_CREDIT'] / df_fe['AMT_INCOME_TOTAL']

# Feature 3: Tuổi của khách hàng (Dữ liệu gốc tính bằng ngày âm, cần chia -365)
df_fe['AGE_YEARS'] = df_fe['DAYS_BIRTH'] / -365

# Feature 4: Tỷ lệ thời gian đi làm / Tuổi đời (Ổn định công việc)
df_fe['EMPLOYED_TO_AGE_RATIO'] = (df_fe['DAYS_EMPLOYED'] / -365) / df_fe['AGE_YEARS']

print("Đã tạo xong các Feature mới. Xem trước 5 dòng:")
display(df_fe.head())
```
